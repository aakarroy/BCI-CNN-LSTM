```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["OPENBLAS_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["VECLIB_MAXIMUM_THREADS"] = "1"
os.environ["NUMEXPR_NUM_THREADS"] = "1"
import pandas as pd
import numpy as np
from keras import Sequential
from keras.layers import LSTM,Dense,Conv1D,BatchNormalization,MaxPool1D,Input,Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
from imblearn.over_sampling import SMOTE

def SMOTE_function(X,Y):
    N,T,C = X.shape
    x_flat = X.reshape(N,T*C)
    smote = SMOTE(random_state=42)
    x_flat,Y = smote.fit_resample(x_flat,Y)
    X = x_flat.reshape(-1,T,C)
    return X,Y

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    y = lfilter(b, a, data, axis=0)
    return y

def slicing_data(cnt_path,mk_path,jump,i):
    cnt = np.loadtxt(cnt_path)
    mk = np.loadtxt(mk_path)
    cnt = butter_bandpass_filter(cnt, 8.0, 30.0, 100.0)
    epochs = []
    labels = []
    event_pos = mk[:,0].astype(int)
    event_type = mk[:,1].astype(int)
    if(i=='a'):
        event_type = np.where(event_type==-1,0,2)
    if(i=='b'):
        event_type = np.where(event_type==-1,0,1)
    if(i=='c'):
        event_type = np.where(event_type==-1,0,1)
    if(i=='d'):
        event_type = np.where(event_type==-1,0,1)
    if(i=='e'):
        event_type = np.where(event_type==-1,0,1)
    if(i=='f'):
        event_type = np.where(event_type==-1,0,2)
    if(i=='g'):
        event_type = np.where(event_type==-1,0,1)
    offset = 50
    for i,pos in enumerate(event_pos):
        start = pos+offset
        end = pos+jump
        if(end<=cnt.shape[0]):
            segment = cnt[start:end,:]
            epochs.append(segment)
            labels.append(event_type[i])
    X = np.array(epochs)
    Y = np.array(labels)
    return X,Y

def compile_data(jump):
    all_epochs = []
    all_labels = []
    for i in ['a','b','c','d','e','f','g']:
        x_temp, y_temp = slicing_data(f"/kaggle/input/berlin-bci-dataset-1/BCICIV_1_asc/BCICIV_calib_ds1{i}_cnt.txt",
                                      f"/kaggle/input/berlin-bci-dataset-1/BCICIV_1_asc/BCICIV_calib_ds1{i}_mrk.txt",jump,i)
        all_epochs.append(x_temp)
        all_labels.append(y_temp)
    X_final = np.concatenate(all_epochs,axis=0)
    Y_final = np.concatenate(all_labels,axis=0)
    return X_final,Y_final

def preprocessing(X,Y):
    scaler = StandardScaler()
    N,T,C = X.shape
    x_flat = X.reshape(N,T*C)
    x_flat = scaler.fit_transform(x_flat)
    return x_flat.reshape(N,T,C),Y


def model_architecture(N,T,C):
    model = Sequential([
        Input(shape = (T,C)),
        Conv1D(16,128,strides=1,activation='relu',padding="same"),
        BatchNormalization(),
        MaxPool1D(pool_size=2),

        Conv1D(32,64,strides=1,activation='relu',padding="same"),
        BatchNormalization(),
        MaxPool1D(pool_size=2),

        Conv1D(64,32,strides=1,activation='relu',padding="same"),
        BatchNormalization(),
        MaxPool1D(pool_size=2),

        Conv1D(128,16,strides=1,activation='relu',padding="same"),
        BatchNormalization(),
        MaxPool1D(pool_size=2),
        
        LSTM(units=256,activation="tanh",dropout=0.1,return_sequences=True),
        
        LSTM(units=128,activation="tanh"),
        
        Dense(256,activation="relu"),

        Dense(128,activation="relu"),
        
        Dropout(0.4),
        
        Dense(3,activation="softmax")
    ])
    model.compile(optimizer="AdamW",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
    return model

fs = 100
epoch_duration = 4
X,Y = compile_data(fs*epoch_duration)
X,Y = preprocessing(X,Y)

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
x_train,y_train = SMOTE_function(x_train,y_train)
N,T,C = x_train.shape
model = model_architecture(N,T,C)
print(model.summary())

history = model.fit(
    x_train,y_train,batch_size=64,epochs=30,validation_split=0.1,verbose=2,shuffle=True
)
loss, accuracy = model.evaluate(x_test,y_test)

print(loss)
print(accuracy)
# print(precision)
# print(recall)

plt.plot(history.history["accuracy"],label="Accuracy")
plt.plot(history.history["val_accuracy"],label="Validation Accuracy")
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
plt.savefig("Model_Accuracy_can_be7.png")
plt.show()

# plt.plot(history.history["precision"],label="Precision")
# plt.plot(history.history["val_precision"],label="Validation Precision")
# plt.title("Model Precision")
# plt.ylabel("Precision")
# plt.xlabel("Epochs")
# plt.legend()
# plt.savefig("Model_Precision_final1.png")
# plt.show()

# plt.plot(history.history["recall"],label="Recall")
# plt.plot(history.history["val_recall"],label="Validation Recall")
# plt.title("Model Recall")
# plt.ylabel("Recall")
# plt.xlabel("Epochs")
# plt.legend()
# plt.savefig("Model_Recall_final1.png")
# plt.show()

plt.plot(history.history['loss'],label="Train Loss")
plt.plot(history.history['val_loss'],label="Val Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend()
plt.savefig("Model_Loss_can_be7.png")
plt.show()

model.save(r"/kaggle/working/BCI_can_be_7.h6")
```
